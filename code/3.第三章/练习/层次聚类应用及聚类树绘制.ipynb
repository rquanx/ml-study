```python
import pandas as pd

df = pd.read_csv("challenge-8-seeds.csv")
df.head()
```

```python
df.shape
```

```python
from scipy.cluster import hierarchy

# 返回一个 Nx4 的矩阵
# [1.72000000e+02, 2.06000000e+02, 1.17378192e-01, 2.00000000e+00] 
# 表示 172 类别和 206 类别被合并，当前距离为 1.17378192e-01 属于全集合最短距离，合并后类别中包含有 2 个数据样本
# 进行层次聚类，结果有 209 类
Z = hierarchy.linkage(df, method="ward")
Z[:5],Z.shape
```

```python
from matplotlib import pyplot as plt

%matplotlib inline

plt.figure(figsize=(15, 8))
hierarchy.dendrogram(Z)
plt.show()
```

```python
# 横线所在高度表明类别间合并时的距离。如果相邻两横线的间距越大，则说明前序类别在合并时的距离越远，也就表明可能并不属于一类不需要被合并
```

```python
plt.figure(figsize=(15, 8))

# 调整参数来修剪聚类树
hierarchy.dendrogram(Z, truncate_mode='lastp', p=15, show_leaf_counts=True, show_contracted=True)
plt.show()
```
