```python
import pandas as pd

df = pd.read_csv("./shopping_data.csv",header=None)

df.values
```

```python
# apply(function, axis=1): 这个方法会对 DataFrame 的每一行（axis=1）应用 
# x.dropna(): 这个方法会移除 x 中的 NaN 值。对于每一行来说，它会移除该行中的所有 NaN 元素。 
# .to_numpy(): 这个方法会将 dropna() 后的结果转换为 NumPy 数组
# to_numpy(): 最后将 apply 的结果（一个包含 NumPy 数组的 Pandas Series）转换为 NumPy 数组
# axis=0：沿着行方向操作（即对每列进行操作）
# axis=1：沿着列方向操作（即对每行进行操作）。
cleaned_data = df.apply(lambda x: x.dropna().to_numpy(), axis=1).to_numpy()

cleaned_data
```

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()  # 定义模型
te_ary = te.fit_transform(cleaned_data)  # 转换数据集
te_ary
```

```python
ddf = pd.DataFrame(te_ary, columns=te.columns_)  # 将数组处理为 DataFrame
ddf
```

```python
from mlxtend.frequent_patterns import apriori


# 预设支持度寻找频繁项集
frequent_itemsets = apriori(ddf, min_support=0.05, use_colnames=True) 
```

```python
from mlxtend.frequent_patterns import association_rules

association_rules(
    frequent_itemsets,  min_threshold=0.2
)  # 置信度阈值为 0.6
```
