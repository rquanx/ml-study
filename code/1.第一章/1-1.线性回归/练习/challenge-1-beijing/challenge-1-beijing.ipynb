```python
import pandas as pd

df = pd.read_csv(
    "./challenge-1-beijing.csv"
)
df.head()
```

```python
features = df[['公交','写字楼','医院','商场','地铁','学校','建造时间','楼层','面积']]
target = df["每平米价格"]
features.head(),target.head()
```

```python

split_num = int(len(features) * 0.7)  # 得到 70% 位置

# :x === 0 ~ x slice
X_train = features[:split_num]  # 训练集特征
y_train = target[:split_num]  # 训练集目标

# x: === x ~ end slice
X_test = features[split_num:]  # 测试集特征
y_test = target[split_num:]  # 测试集目标

X_train.shape,y_train.shape, X_test.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression


model = LinearRegression()
model.fit(X_train, y_train)

# 得到模型拟合参数
model.intercept_, model.coef_
```

```python
preds = model.predict(X_test)  # 输入测试集特征进行预测
preds  # 预测结果
```

```python
import numpy as np

def mape_slover(y_true: np.ndarray, y_pred: np.ndarray):
    """mape 求解"""
    n = len(y_true)
    return np.sum(np.abs((y_true - y_pred) / y_true)) / n * 100 

mape = mape_slover(y_test.values,preds)
print(f"mape: {mape}%")
```
